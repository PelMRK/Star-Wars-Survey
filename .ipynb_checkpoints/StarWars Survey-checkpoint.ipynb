# Star Wars Survey

## Overview

While waiting for [Star Wars: The Force Awakens](https://en.wikipedia.org/wiki/Star_Wars:_The_Force_Awakens) to come out, the team at [FiveThirtyEight](https://fivethirtyeight.com) became interested in answering some questions about *Star Wars* fans. In particular, they wondered: **does the rest America realize that "The Empire Strikes Back" is clearly the best of the bunch?**

The team needed to collect data addressing this question. To do this, They surveyed *Star Wars* fans using the online tool SurveyMonkey. They received 835 total responses, which you can download from [their GitHub repository](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey).

For this project we'll be cleaning and exploring the data set.

The data has several columns, including:

- `RespondentID` - An anonymized ID for the respondent (person taking the survey)


- `Gender` - The repondent's gender


- `Age` - The respondent's age


- `Household Income` - The respondent's income


- `Education` - The respondent's education level


- `Location (Census Region)` - The respondent's location


- `Have you seen any of the 6 films in the Star Wars franchise?` - Has a `Yes` and `No` response


- `Do you consider yourself to be a fan of the Star Wars film franchise?` - Has a `Yes` and `No` response


There are several other columns containing answers to questions about the *Star Wars* movies. For some questions, the respondent had to check one or more boxes. This type of data is difficult to represent in columnar format. As a result, this data set needs a lot of cleaning.

```python
import pandas as pd

star_wars = pd.read_csv("StarWars.csv", encoding="ISO-8859-1")  # contains characters that aren't in default utf-8
star_wars = star_wars[1:]
star_wars.head(10)
```

```python
star_wars.columns
```

## Cleaning and Mapping Yes/No Columns

We'll convert `Yes` and `No` answers to Boolean that we can make the data a bit easier to analyze.

```python
mask = {"Yes": True, "No": False}

for col in ["Have you seen any of the 6 films in the Star Wars franchise?",
            "Do you consider yourself to be a fan of the Star Wars film franchise?"]:
    star_wars[col] = star_wars[col].map(mask)

star_wars.head()
```

## Cleaning and Mapping Checkbox And Ranking Columns

The next six columns represent a simple checkbox question. For each of these columns, if the value in a cell is the name of the movie, that means the respondent saw the movie. If the value is `NaN`, the respondent either didn't answer or didn't see the movie. We'll assume that they didn't see the movie.

And the next six columns ask respondent to rank the *Star Wars* movies in order of least favorite to most favorite. `1` means the film was the most favorite, and `6` - the least favorite.

Next, we'll need to rename the columns to better reflect what they represent. 

```python
import numpy as np

star_wars.iloc[:,3:9] = star_wars.iloc[:, 3:9].applymap(lambda x: x is not np.nan)
star_wars.iloc[:,9:15] = star_wars.iloc[:, 9:15].astype(float)

mask = {}
for i, name in enumerate(star_wars.columns[3:9]):
    mask[name] = "seen_" + str(i+1)
    
for i, name in enumerate(star_wars.columns[9:15]):
    mask[name] = "ranking_" + str(i+1)
    
star_wars = star_wars.rename(columns=mask)
star_wars.head()
```

## Finding the Highest-Ranked Movie

Now that we've cleaned up the ranking columns, we can find the highest-ranked movie. 

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.title('Movie Ranking')
plt.ylim(2, 4.5)
plt.bar(range(1, 7), star_wars.iloc[:, 9:15].mean())
```

Remember that a lower ranking is better here. For some reason, respondents put the latest movies higher, although it was expected that the "original" movies would be rated much more highly than the newer ones.

## Finding the Most Viewed Movie

We can figure out how many people have seen each movie.

```python
plt.title('Movie Views')
plt.ylim(400, 800)
plt.bar(range(1, 7), star_wars.iloc[:, 3:9].sum())
```

Looks like the majority of respondents have only watched the latest films, which may be reason why they rate them higher.

## Exploring the Data by Binary Segments

We know which movies the survey population as a whole ranked the highest. Now let's examine how certain segments of the survey population responded. There are several columns that segment our data into two groups:

- Considering to be the Star Wars fan


- Considering to be the Star Trek fan


- Gender

### Gender

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.ylim(2, 4.5)
plt.title('Male Movie Ranking')
plt.bar(range(1, 7), males.iloc[:, 9:15].mean())

plt.subplot(2, 2, 2)
plt.ylim(2, 4.5)
plt.title('Female Movie Ranking')
plt.bar(range(1, 7), females.iloc[:, 9:15].mean())

plt.subplot(2, 2, 3)
plt.ylim(200, 400)
plt.title('Male Movie Views')
plt.bar(range(1, 7), males.iloc[:, 3:9].sum())

plt.subplot(2, 2, 4)
plt.ylim(200, 400)
plt.title('Female Movie Views')
plt.bar(range(1, 7), females.iloc[:, 3:9].sum())

plt.show()
```

More males watched episodes 1-3, but males liked them far less than females did.

### Star Wars Fans

```python
sw_fans = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == True]
not_sw_fans = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == False]

plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.ylim(2, 4.5)
plt.title('SW Fans Movie Ranking')
plt.bar(range(1, 7), sw_fans.iloc[:, 9:15].mean())

plt.subplot(2, 2, 2)
plt.ylim(2, 4.5)
plt.title('Not SW Fans Movie Ranking')
plt.bar(range(1, 7), not_sw_fans.iloc[:, 9:15].mean())

plt.subplot(2, 2, 3)
plt.ylim(0, 600)
plt.title('SW Fans Movie Views')
plt.bar(range(1, 7), sw_fans.iloc[:, 3:9].sum())

plt.subplot(2, 2, 4)
plt.ylim(0, 600)
plt.title('Not SW Fans Movie Views')
plt.bar(range(1, 7), not_sw_fans.iloc[:, 3:9].sum())

plt.show()
```

Star Wars fans had watched movies much more than non-fans. Their views distribution is pretty uniform. Also, Star Wars fans rated newer film much better, whilst non-fans preferred "classic" old films.

### Star Trek Fans

```python
st_fans = star_wars[star_wars['Do you consider yourself to be a fan of the Star Trek franchise?'] == 'Yes']
not_st_fans = star_wars[star_wars['Do you consider yourself to be a fan of the Star Trek franchise?'] == 'No']

plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.ylim(2, 4.5)
plt.title('ST Fans Movie Ranking')
plt.bar(range(1, 7), st_fans.iloc[:, 9:15].mean())

plt.subplot(2, 2, 2)
plt.ylim(2, 4.5)
plt.title('Not ST Fans Movie Ranking')
plt.bar(range(1, 7), not_st_fans.iloc[:, 9:15].mean())

plt.subplot(2, 2, 3)
plt.ylim(0, 600)
plt.title('ST Fans Movie Views')
plt.bar(range(1, 7), st_fans.iloc[:, 3:9].sum())

plt.subplot(2, 2, 4)
plt.ylim(0, 600)
plt.title('Not ST Fans Movie Views')
plt.bar(range(1, 7), not_st_fans.iloc[:, 3:9].sum())

plt.show()
```

The behavior of Star Trek fans and non-fans slighlty differs. The fans prefer newer films a bit more often.

## Conclusion

We have cleaned and explored dataset and found some patterns:
- The newer movies are watched more often and rated higher than the older


- More males watched episodes 1-3, but males liked them far less than females did


- Star Wars fans had watched movies much more than non-fans and rated newer film much better


- The behavior of Star Trek fans and non-fans slighlty differs
